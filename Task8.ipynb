# Anime
Anime, or Japanese animation, is a form of media that started to air in the early 20th century with the first movie being Namakura Gatana (The Dull Sword), a short silent film created by Jun'ichi Kōuchi in 1917. Anime wasn't much in the beginning, but it started to gain a bit more attention after World War II. Since then, it has grown to be a huge success with recent movies beating Hollywood at the box office. It is a piece of content that is usually adapted after the success of a manga(Japanese comic) or light novel. Recently, <b> One Piece </b> surpassed sales of over 600 million copies globally, surpassing Superman and Batman sales, making it a global phenomenon.
<br> The main reason for the success of anime is their storytelling, art style and huge variety of genres to pick shows from.
### Anime Genres
There are over 40 genres that are popular and well recognised, with a few of the main names being
1) Isekai-> genre of fiction where a protagonist is transported, reincarnated, or summoned to a different world—often a fantasy realm, parallel universe, or virtual world. eg-That Time I Got Reincarnated as a Slime, Sword Art Online.
2) Shonen->   class of manga for an audience of young boys. eg- One Piece, Naruto
3) shojo->  class of manga for an audience of young girls. eg-Fruits Basket, Violet Evergarden
4) Seinen: For adult men (18+), often with mature themes like violence, psychological depth, and complex narratives. Examples: Berserk, Monster, Vinland Saga
5) Action: Fast-paced, fight-centric, with intense battles. Examples: Demon Slayer, Attack on Titan.
6) Adventure: Focuses on exploration, quests, and discovery. Examples: One Piece, Hunter x Hunter.
7) Comedy: Lighthearted, humorous, often situational. Examples: One-Punch Man, The Devil Is a Part-Timer!.
8) Slice of Life: Depicts everyday routines and relationships. Examples: Yuru Camp, Natsume’s Book of Friends.
9) Romance: Centres on love and relationships. Examples: Clannad, Your Lie in April.
10) Mecha: Involves giant robots (mechs) and their pilots. Examples: Code Geass, Neon Genesis Evangelion.
11) Psychological: Explores mental states, morality, and inner conflict. Examples: Death Note, Psycho-Pass.
12) Sci-Fi: Features futuristic technology, space travel, or advanced societies. Examples: Steins Gate, Gintama.
13) Fantasy: Set in magical or supernatural worlds. Examples: Spirited Away, Little Witch Academia
